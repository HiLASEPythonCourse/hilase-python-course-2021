### Indexing vs. copying

Indexing in `numpy` does not have to create copies. Data copy is created only when inevitable.

```python
original_array = np.linspace(0, 1, 3)
original_array
```

```python
array_slice = original_array[1: ]
array_slice *= 2
array_slice
```

## Vector operations

Numpy is optimized for performance. One of the key ingredient are vectorized operations: Instead of (slow) looping by elements, an operation can be performed on the whole or at least a subset of an array.

Let's try to implement a dot product (1D for simplicity):

```python
def dot(x, y):
    result = 0
    for i, j in zip(x, y):
        result += i * j
    
    return result
```

```python
x = np.random.sample((4, ))
y = np.random.sample((4, ))

dot(x, y)
```

Test our function using `allclose` to compare floats only up to certain precision.

```python
assert np.allclose(dot(x, y), x.dot(y))
```

We can compare the performance (on larger arrays)

```python
size = (1000, )
x_large = np.random.sample(size)
y_large = np.random.sample(size)
```

```python
%timeit dot(x_large, y_large)
```

```python
%timeit x_large.dot(y_large)
```

Even multiplication (`*` is element-wise in Numpy) is vectorized. E.g.:

```python
x * y
```

## Aggregation

Aggregate function turn multiple values (vector elements, columns, matrices, ...) into a single value.
Typical examples are minimum, sum, product or mean. 
Boolean `all` or `any`, which we have seen already, are also aggregations.

In Numpy, aggregation is performed as an `ndarray` method call or via a function call.

Aggregations in Numpy operate on whole arrays by default, for 1-D (vectors), 2-D (matrices) or more-D arrays
![](images/numpy/numpy-array-aggregation.png)

## Resize & reshape

Sometimes, the elements of your arrays are ok, but their shape is not
(e.g. you read a linear series of pixel values, knowing they form
a two-dimensional image with three channels). To reinterpret the data
with a new shape, use the [`reshape`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.reshape.html) method:



```python
data = np.arange(1, 7)
data
```

![](images/numpy/numpy-reshape.png)

```python
data.reshape(2, 3)
```

```python
data.reshape(3, 2)
```

Note: `reshape` just takes the data as a linear array and applies the shape on it.
So when reshaping an array multiple times, only "the last counts":

```python
data.reshape(3, 2).reshape(2, 3)
```

If you are lazy to calculate the exact shape, you may omit one of the dimensions and give `-1`
in its place - it gets computed for you:

```python
data.reshape(3, -1)  # "Guesses" the right dim is 2
```

You might have spotted that the two 2x3 arrays are not transpose of each other.
To do so, you may want to use the [`transpose`](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.transpose.html)
method (or equivalently, the `T` property):

```python
data.reshape(3, -1).transpose()
```

```python
data.reshape(3, -1).T
```

## Broadcasting

Numpy automatically broadcasts (aligns / enhances the dimensions) of arrays if it "makes sense" 
(there are strict rules :). Broadcasting simplifies operations and can make them faster 
and / or less memory consuming.
The memory consumption can, however, increase when using broadacasting as well.
This is because short-lived helper arrays are created in the memory during broadcasting.

An example is to substract the column mean from an $n \times m$ array.

```python
X = np.linspace(0, 1, 15).reshape(-1, 3)
X
```

```python
X.mean(axis=0)
```

Now we'd like to center the columns around zero, i.e. make the mean = 0 of all columns.

```python
X_centered = X - X.mean(axis=0)
```

```python
X_centered.mean(axis=0)
```

We can also do this explicitely with `newaxis`

```python
X_centered = X - X.mean(axis=0)[np.newaxis, :]
```

## Under the hood: the buffer protocol

`ndarray` data are stored in (typically) contiguous memory blocks. This is for performance reasons because Numpy can employ fast C libraries (like BLAS) for vectorized operations without converting and / or rearranging the data in the memory. Python offers built-in [Buffer Protocol](https://docs.python.org/3/c-api/buffer.html#bufferobjects) which standardizes this approach and enables (relatively) easy integration with C libraries.

In a nutshell, `ndarray` data are described by their memory buffer location (pointer), and some metedata mainly describing the data type (`typestr`) and dimensionality (`shape` and `strides`).

Looking a bit under the hood, we can see the buffer info by `info` or `flags`

```python
np.info(array)
```

```python
array.flags
```

This is an implementation detail, although very important because it determines fundamental advantages, limitations and best practices for using `numpy`.
